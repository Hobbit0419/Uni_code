```python
import numpy as np
from matplotlib import pyplot as pp
```

Uppgift 1.2

```python
def f(x):
    return np.exp(-x**2)

def P_1(h):
    return np.e**-1 - 2*h*np.e**-1

def P_2(h):
    return np.e**-1 - 2*h*np.e**-1 + 2*h**2*np.e**-1

h = np.arange(0, 1, 0.01)

err_1 = np.abs(P_1(h) - f(h+1))
err_2 = np.abs(P_2(h) - f(h+1))

y = f(h+1)
y_1 = P_1(h)
y_2 = P_2(h)
```

```python
fig, ax = pp.subplots()
ax.plot(h, y, label='$f(x_{i+1})$')
ax.plot(h, y_1, label='$P_1(x_{i+1})$')
ax.plot(h, y_2, label='$P_2(x_{i+1})$')
ax.legend()
```

För små värden på h är både p1 och p2 bra approximationer men när h överstiger 0.2 har både p1 och p2 avikit från f ganska rejält dock på varsitt håll.

```python
fig, ax = pp.subplots()

ax.plot(h, err_1, label='$|P_1(h) - f(1+h)|$')
ax.plot(h, err_2, label='$|P_2(h) - f(1+h)|$')
ax.legend()
```

Ur granfen ovan kan man ganska lätt se att p2 är en bättre approximation då dess fel ökar långsammare.
