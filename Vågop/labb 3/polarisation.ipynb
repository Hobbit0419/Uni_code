```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
theta = np.linspace(-90,90,19)
I = np.array([0.3,0.6,1.8,3.3,4.9,6.4,7.5,8.5,9.1,9.4,9.4,8.8,7.9,6.8,5.5,4.0,2.3,1.0,0.3])
It = 9.4 * np.cos(np.deg2rad(theta))**2 #Transmiterat värde använt för I0

plt.scatter(theta, I, label='Experimentell')
plt.scatter(theta, It, label='Teoretisk')
plt.legend()
```
