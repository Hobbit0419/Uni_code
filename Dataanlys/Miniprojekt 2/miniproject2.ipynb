```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load training data
train_data = np.load('HandwrittenDigits/TrainDigits.npy')
train_labels = np.load('HandwrittenDigits/TrainLabels.npy')

#Load test data
test_data = np.load('HandwrittenDigits/TestDigits.npy')
test_labels = np.load('HandwrittenDigits/TestLabels.npy')

print(test_labels)
```

```python
#Show the first 10 images
fig, ax = plt.subplots(1, 5, figsize=(5, 2))
for i in range(5):
    img = train_data[:, i]
    img = img.reshape((28, 28)).T
    ax[i].imshow(img, cmap='gray')
    ax[i].axis('off')
```

```python
# Digit 3
N = 400 # number of training samples

index3 = np.where(train_labels[0] == 3)[0] # find all the samples with label 3
Tmatrix3 = train_data[:, index3[:N]] # extract the first N samples with label 3

# Compute the SVD
U3, S3, VT3 = np.linalg.svd(Tmatrix3)

# Plot the singular values
fig1, ax1 = plt.subplots()
ax1.plot(S3/S3[0], color='firebrick')
ax1.set_title('Singular values of the matrix of digit 3')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value (normalized)')
```

```python
# Singular images
fig2, ax2 = plt.subplots(1, 3, figsize=(10, 3))
for i in range(3):
    img = U3[:,i].reshape((28, 28)).T
    ax2[i].imshow(img, cmap='gray')
    ax2[i].axis('off')
fig2.suptitle('First 3 singular images of digit 3')
```

```python
# Digit 8
N = 400 # number of training samples

index8 = np.where(train_labels[0] == 8)[0] # find all the samples with label 8
Tmatrix8 = train_data[:, index8[:N]] # extract the first N samples with label 8
print(Tmatrix8.shape)

# Compute the SVD
U8, S8, VT8 = np.linalg.svd(Tmatrix8)

# Plot the singular values
fig2, ax2 = plt.subplots()
ax2.plot(S8/S8[0], color='firebrick')
ax2.set_title('Singular values of the matrix of digit 8')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value(normalized)')
```

```python
# Singular images
fig3, ax3 = plt.subplots(1, 3, figsize=(10, 3))
for i in range(3):
    img = U8[:,i].reshape((28, 28)).T
    ax3[i].imshow(img, cmap='gray')
    ax3[i].axis('off')
fig3.suptitle('First 3 singular images of digit 8')
```
