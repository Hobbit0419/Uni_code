```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([9,10,11,12,13])
y = np.array([5.0,5.0,4.0,3.0,1.0])

x_fit = np.linspace(9,13,100)

plt.scatter(x,y, label='Data',color='maroon')
plt.plot(x_fit, 14.6+(-1*x_fit), label='y = 14.6 - x')
plt.legend()
```

```python
a = np.array([[615,55],[55,5]])
b = np.array([[10,0],[0,5]])

print(np.linalg.cond(a), np.linalg.cond(b))
```

```python
A = np.array([[9,1],[10,1],[11,1],[12,1],[13,1]])

print(np.shape(A))

Q,R = np.linalg.qr(A, mode='reduced')


print(f'A = \n{A}')
print(f'Q = \n{Q}')
print(f'R = \n{R}')
```

```python
sol = np.linalg.solve(R, np.dot(Q.T,y))

print(f'solution = {sol}')
```

```python
A1 = np.array([[2,1,3],[0,3,-1],[0,1,1],[0,2,-4]])

x = np.array([3,1,2])

v = x + np.linalg.norm(x)*np.array([1,0,0])

H = np.identity(3) - (2/(np.inner(v,v))) * np.outer(v,v)

matrix_4x4 = np.zeros((4, 4))

matrix_4x4[0, 0] = 1

matrix_4x4[1:, 1:] = H

H = matrix_4x4

A2 = np.dot(H, A1)

print(f'v = {v.round(3)}')
print(f'A1 = \n{A1}')
print(f'H = \n{H.round(3)}')
print(f'A2 = \n{A2.round(3)}')
```
