```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import simpson
```

```python
# Wavefunction

def psi(x,a):
    c = (1/a)*np.sqrt(2/np.pi)
    return c*(a**2-x**2)**(0.25)

# Probability density
def prob(x,a):
    return np.abs(psi(x,a))**2
```

```python
a = 1
x = np.linspace(-a,a,1000)

sigma = np.sqrt(simpson(x**2*prob(x,a),x))

print(f'sigma = {sigma.round(2)}a')

fig, ax = plt.subplots()
ax.plot(x, psi(x,a), label='Wave function')
ax.plot(x, prob(x,a), label='Probability density')
ax.vlines([-3*sigma, -2*sigma, -sigma, sigma, 2*sigma, 3*sigma], 0, 1, linestyles='dashed')
ax.legend()
```

```python
a = 1
x = np.linspace(-a,a,1000)

fig, ax = plt.subplots()
ax.plot(x, psi(x,a)/psi(0,a), label='Wave function')
ax.plot(x, prob(x,a)/prob(0,a), label='Probability density')
ax.legend()
```
