```python
import numpy as np
import matplotlib.pyplot as plt
import cmath as cm
from scipy import integrate
```

```python
pointsx = (-10,-6,-2,2,6,10)
pointsy = (3,2,1,0,1,0)

x = np.linspace(-10, 10, 1000)
y = np.interp(x, pointsx, pointsy, left=0, right=0)
integral = integrate.trapezoid(y, x)
y = y/integral
plt.plot(x, y, label='Interpolated Line')

mean = integrate.trapezoid(y*x, x)
variance = integrate.trapezoid(y*(x-mean)**2, x)
stddev = np.sqrt(variance)
print("Mean: ", mean)
print("Variance: ", variance)
print("Standard Deviation: ", stddev)
```

```python
def wave_func(x,A=2):
    return A*(x**2)*np.exp(-np.abs(x)+2j*x)

x = np.linspace(-10, 10, 10000)
y = wave_func(x)
plt.plot(x, y.real, label='Real Part')
plt.plot(x, np.sqrt(y*np.conj(y)), label='Magnitude')
```
