```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
wavelength = np.array([365,405,436,546,577])
frequency = 2.99792458e8 / (wavelength*1e-9)
stopping_potential = np.array([-1.80,-1.41,-1.18,-0.67,-0.55])
```

```python
coefficients, cov = np.polyfit(frequency, stopping_potential, 1, cov=True)
slope, intercept = coefficients
uncertainty = np.sqrt(np.diag(cov))
h = slope * -1.602e-19
h_0 = 6.62607015e-34
dh = np.abs((h_0 - h)/h_0)*100
print(f"Planck's constant: {round(h*1e34,2)} ± {round(uncertainty[0]*1e34*1.602e-19,1)}e-34 J*s")
print(f"Relative error: ±{round(dh,2)}%")
print(f"Slope: {round(slope*1e15,2)} ± {round(uncertainty[0]*1e15,1)}e-15 V/Hz")



fig, ax = plt.subplots()
ax.scatter(frequency, stopping_potential, label='Data', color='firebrick')
ax.plot(frequency, slope * frequency + intercept, color='maroon', label='Fit')
ax.legend()
ax.set_xlabel('Frequency (Hz)')
ax.set_ylabel('Stopping Potential (V)')
ax.set_title('Stopping Potential vs Frequency')
```
