## 1d)
    Illustrera sannolikhetsfunktionen som ett stapeldiagram med lämplig stapelbredd och lägg i samma graf in en sannolikhetsfunktion från Poisson-fördelningen med någon lämplig parameter. Diskutera kort om det här verkar vara en bra approximation eller ej.   

```python
def primes(n):
    factors = []
    d = 2
    while d <= n:
        if n % d == 0:
            factors.append(d)
            n = n / d
        else:
            d += 1
    return factors
```
