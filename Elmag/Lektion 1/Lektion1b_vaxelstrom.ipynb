sinusformade sp√§nningar och str√∂mmar

Fr√•n
\begin{align}
ùë£(ùë°)=\hat{ùëâ}¬†\sin{(\omega ùë° + \phi)}‚Å°
\end{align}

d√§r
\begin{align}
ùë£(ùë°) & \quad \text{momentantv√§rde}\\
\omega = 2\pi ùëì & \quad \text{vinkelfrekvens}\\
\phi & \quad \text{fasvinkel}\\
\hat{ùëâ} & \quad \text{toppv√§rde}
\end{align}
man g√•r √∂ver till vektornotation, man kallar de dock inte f√∂r vektorer men **visare** (**phasors**)

### Visarform:

$\bar{ùëâ} = ùëâ ùëí^{ùëó\phi} = ùëâ \angle \phi$ d√§r $V$  √§r **effektivv√§rde**

\begin{align}
V = \sqrt{\frac{1}{T} \int _{0}^{T} v(t)^2 dt}\\
\text{F√∂r sinusformade signaler g√§ller}\\
V = V_{ùëüùëöùë†} = \hat{V}/\sqrt{2}
\end{align}

### Exempel: 
#### Eln√§tets periodiska 50 Hz v√§xelsp√§nning. V√§xelsp√§nningens effektivv√§rde √§r 230 V - toppv√§rdet √§r 325 V




---

```python
%matplotlib inline
import cmath as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = 10 # sp√§nningstoppv√§rde
I = 8 # str√∂mmenstoppv√§rde
f = 1000 # signalfrekvens 1kHz
phi1 = 0 # fas f√∂r signal 1
phi2 = -np.pi/4 # fas for signal 2
N = 1000 # antal punkter 
T = 1/f # periodtid
tMax = 3*T 
t = np.linspace(0,tMax, N) # tids axel
w = (2*np.pi)*f # vinkelfrekvens
```

### Skapar sinusformade signaler f√∂r sp√§nning och str√∂m

```python
v = V*np.sin(w*t + phi1)
i = I*np.sin(w*t + phi2)
```

### Ritar sp√§nning och str√∂m tillsammans

```python
fig1, ax1 = plt.subplots()
x = t*1e3
ax1.plot(x, v,'b', x, i,'r')
ax1.set_xlabel('Time [ms]')
ax1.set_ylabel('Voltage [V]')
ax1.grid(True)
```

### Ritar samma signaler uttryckt som visorer

```python
fig2, ax2 = plt.subplots(subplot_kw=dict(projection="polar"))
xPhasor1 = [0, phi1]
yPhasor1 = [0, V];
xPhasor2 = [0, phi2]
yPhasor2 = [0, I];
ax2.plot(xPhasor1, yPhasor1,'b');
ax2.plot(xPhasor2, yPhasor2,'r');
```
