## **Växelström** 
sinusformade spänningar och strömmar

Från
\begin{align}
𝑣(𝑡)=\hat{𝑉} \sin{(\omega 𝑡 + \phi)}⁡
\end{align}

där
\begin{align}
𝑣(𝑡) & \quad \text{momentantvärde}\\
\omega = 2\pi 𝑓 & \quad \text{vinkelfrekvens}\\
\phi & \quad \text{fasvinkel}\\
\hat{𝑉} & \quad \text{toppvärde}
\end{align}
man går över till vektornotation, man kallar de dock inte för vektorer men **visare** (**phasors**)

### Visarform:

$\bar{𝑉} = 𝑉 𝑒^{𝑗\phi} = 𝑉 \angle \phi$ där $V$  är **effektivvärde**

\begin{align}
V = \sqrt{\frac{1}{T} \int _{0}^{T} v(t)^2 dt}\\
\text{För sinusformade signaler gäller}\\
V = V_{𝑟𝑚𝑠} = \hat{V}/\sqrt{2}
\end{align}

### Exempel: 
#### Elnätets periodiska 50 Hz växelspänning. Växelspänningens effektivvärde är 230 V - toppvärdet är 325 V




---

```python
%matplotlib inline
import cmath as cm
import matplotlib.pyplot as plt
import numpy as np
```

```python
V = 10 # spänningstoppvärde
I = 8 # strömmenstoppvärde
f = 1000 # signalfrekvens 1kHz
phi1 = 0 # fas för signal 1
phi2 = -np.pi/4 # fas for signal 2
N = 1000 # antal punkter 
T = 1/f # periodtid
tMax = 3*T 
t = np.linspace(0,tMax, N) # tids axel
w = (2*np.pi)*f # vinkelfrekvens
```

### Skapar sinusformade signaler för spänning och ström

```python
v = V*np.sin(w*t + phi1)
i = I*np.sin(w*t + phi2)
```

### Ritar spänning och ström tillsammans

```python
fig1, ax1 = plt.subplots()
x = t*1e3
ax1.plot(x, v,'b', x, i,'r')
ax1.set_xlabel('Time [ms]')
ax1.set_ylabel('Voltage [V]')
ax1.grid(True)
```

### Ritar samma signaler uttryckt som visorer

```python
fig2, ax2 = plt.subplots(subplot_kw=dict(projection="polar"))
xPhasor1 = [0, phi1]
yPhasor1 = [0, V];
xPhasor2 = [0, phi2]
yPhasor2 = [0, I];
ax2.plot(xPhasor1, yPhasor1,'b');
ax2.plot(xPhasor2, yPhasor2,'r');
```
