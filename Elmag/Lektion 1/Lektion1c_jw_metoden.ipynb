## $j\omega$-metoden: sammanfattning

Vid analys av v√§xelstr√∂mskretsar med $j\omega$-metoden g√§ller f√∂ljande:

1. Str√∂mmar och sp√§nningar representeras med komplexa effektivv√§rden (amplituder) varvid den tidsberoende faktorn $ùëí^{ùëó\omega t}$ redan fr√•n b√∂rjan kan slopas.
2. Impedanser representeras med komplexa tal, varvid en ren resistans s√§tts lika med $ùëÖ$, en ren induktans s√§tts lika med $ùëó\omega ùêø$, en ren kapacitans s√§tts lika med $\dfrac{1}{j\omega C}$.
3. Anv√§ndningen av komplexa uttryck f√∂r str√∂mmar, sp√§nningar och impedanser g√∂r det m√∂jligt att behandla v√§xelstr√∂msproblem som om det vore ett likstr√∂msproblem, dvs samtliga str√∂mkretslagar f√∂r likstr√∂m g√§ller.

Exempelvis g√§ller f√∂r seriekopplade impedanser: $ \bar{Z} = \bar{Z_1} + \bar{Z_2} + \bar{Z_3} + ... $

F√∂r parallellkopplade impedanser g√§ller: $ \frac{1}{\bar{Z}} = \frac{1}{\bar{Z_1}} +\frac{1}{\bar{Z_2}} + \frac{1}{\bar{Z_3}} + ... $

Speciellt g√§ller f√∂r tv√• parallellkopplade impedanser: $ \bar{Z} = \frac{\bar{Z_1} \bar{Z_2}}{\bar{Z_1} + \bar{Z_2}} $

Den storhet (str√∂m, sp√§nning, etc) man √∂nskar ber√§kna erh√•lls i slutresultatet i komplex form, $ \bar{A} = a + jb $

Beloppet respektive fasvinkeln (argumentet) r√§knas med $ A = \sqrt{a^2 + b^2}$, och $ arg‚Å°(ùê¥) = \arctan‚Å°(ùëè‚ÅÑùëé)$

#### R√§kna i Python:
- $\bar{I} = 0,4 + 1,2j = 1,3 \angle72^o$
- $\bar{U_1} = 12 - 4j = 13 \angle -18^o$
- $\bar{U_2} = 8 + 4j = 9 \angle 26^o$

- #### Hitta svaret i rektangul√§ra och pol√§ra koordinater.
- #### Rita de signalerna tillsammans med ing√•ngssp√§nningen $U$ i visarform. Du kan beh√∂va skala beloppet av visa signaler s√• att alla syns tydligt p√• samma graf.
- #### Tolka resultat. Vad √§r fas skillnad mellan $\bar{I}$ och  $\bar{U_1}$ respektive  $\bar{I}$ och  $\bar{U_2}$? Varf√∂r?
---

Det √§r $90^o$ mellan $\bar{I}$ och $\bar{U_1}$, d√§r har vi bara en ren kondensator som ger fasf√∂rskjutningen. Mellan $\bar{I}$ och $\bar{U_2}$ √§r det mindre √§n $90^o$ eftersom vi har en total impedans som som inneh√•ller b√•de resistans och kapacitans.


```python
%matplotlib inline
import cmath as cm
import matplotlib.pyplot as plt
import numpy as np

U = 20
C = 320 * 10**-6
R = 10
f = 50
w = f * 2*np.pi

Z1 = complex(0, -1/(w*C))
Z2 = complex(R, 0)
Z3 = complex(0, -1/(w*C))

Zt = Z1 + 1/((1/Z2) + (1/Z3))

I = U / Zt

U1 = Z1*I

U2 = (1/((1/Z2) + (1/Z3))) * I

print(f'I = {I} U1 = {U1} U2 = {U2}')
print(f'I = ({abs(I)},{np.rad2deg(cm.phase(I))}) U1 = ({abs(U1)}, {np.rad2deg(cm.phase(U1))}) U2 = ({abs(U2)},{np.rad2deg(cm.phase(U2))})')

fig1, ax1 = plt.subplots(subplot_kw=dict(projection="polar"))
xPhasor1 = [0, cm.phase(U)]
yPhasor1 = [0, abs(U)/10]
xPhasor2 = [0, cm.phase(U1)]
yPhasor2 = [0, abs(U1)/10]
xPhasor3 = [0, cm.phase(I)]
yPhasor3 = [0, abs(I)]
xPhasor4 = [0, cm.phase(U2)]
yPhasor4 = [0, abs(U2)/10]
ax1.plot(xPhasor1, yPhasor1,'r')
ax1.plot(xPhasor2, yPhasor2,'g')
ax1.plot(xPhasor3, yPhasor3,'b')
ax1.plot(xPhasor4, yPhasor4,'black')
```

### En spole med induktansen 19,1 mH och resistansen 8,00 ‚Ñ¶ √§r parallellkopplad med en kondensator vars kapacitans √§r 318 ¬µF. √ñver denna krets kopplas en sp√§nningsk√§lla med sp√§nningen 230 V och frekvensen 50 Hz. 
 
- #### Ber√§kna str√∂mmen i den gemensamma ledningen ($I$) samt b√•da str√∂mmarna i parallella grenar ($I_C$ och $I_{LR}$). Hitta svaret i rektangul√§ra och pol√§ra koordinater.

$I = 18+9,2j = 21 \angle 27^o$

$I_c = 23j = 23 \angle 90^o$

$I_{LR} = 18-14j = 23 \angle -37^o$

- #### Rita de signalerna tillsammans med ing√•ngssp√§nningen $U$ i visarform. Du kan beh√∂va skala beloppet av visa signaler s√• att alla syns tydligt p√• samma graf.
- #### Tolka resultat. Vad √§r fas skillnad mellan $\bar{I}$ och  $\bar{U}$ respektive  $\bar{I_{LR}}$ och  $\bar{U}$? Varf√∂r? Vad skulle h√§nda med fas skillnader om kapacitansen minskas 10 g√•nger? Testa och kommentera.

##### Vi har $27^o$ mellan U och I samt $-37^o$ mellan U och $I_{LR}$ om vi minskar kapacitansen med 10 g√•nger minskar fasf√∂rskjutningen och den n√§rmar sig de $-37^o$ graderna vi har mellan U och $I_{LR}$. Fasf√∂rskjutningen fr√•n spolen kommer fr√•n en kombination av resistans och impedans vilket ger den fasf√∂rskjutningen vi f√•r, medans den rena kapacitancen fr√•n kondensatorn ger en fasf√∂rskjutning p√• $90^o$ grader vilket spolen ocks√• hade gjort om den var utan resistans. 
---

```python
from phasors import phasor_diagram_rect

L = 19.1 * 10**-3
Rl = 8
C = 318 * 10**-6
U = 230
f = 50
w = 2*np.pi*f

Zc = complex(0, -1/(w*C))
Zl = complex(Rl, w*L)
Zt = 1/((1/Zc)+(1/Zl))

I = U/Zt
Ic = U/Zc
Ilr = U/Zl

print(f'I = {I} Ic = {Ic} Ilr = {Ilr}')
print(f'I = ({abs(I)},{np.rad2deg(cm.phase(I))}) Ic = ({abs(Ic)}, {np.rad2deg(cm.phase(Ic))}) Ilr = ({abs(Ilr)},{np.rad2deg(cm.phase(Ilr))})')

fig2, ax2 = plt.subplots(subplot_kw=dict(projection="polar"))
xPhasor1 = [0, cm.phase(U)]
yPhasor1 = [0, abs(U)/10]
xPhasor2 = [0, cm.phase(I)]
yPhasor2 = [0, abs(I)]
xPhasor3 = [0, cm.phase(Ic)]
yPhasor3 = [0, abs(Ic)]
xPhasor4 = [0, cm.phase(Ilr)]
yPhasor4 = [0, abs(Ilr)]
ax2.plot(xPhasor1, yPhasor1,'r', label='U')
ax2.plot(xPhasor2, yPhasor2,'g', label='I')
ax2.plot(xPhasor3, yPhasor3,'b', label='Ic')
ax2.plot(xPhasor4, yPhasor4,'black', label='Ilr')
ax2.legend()
```
