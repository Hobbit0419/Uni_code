#### Koden nedan räknar och ritar amplituden av totala impedansen i funktion av frekvens i område: 0 – 1000 Hz

### Din uppgift är att räknar och ritar amplituden av strömmen $I$ genom kretsen i funktion av frekvens i område: 0 – 1000 Hz och tolka resultat. Vad händer med strömmen och impedansen?

---

```python
# variera frekvens
f = np.linspace(1,1000,200)
w_a = 2*np.pi*f
V0 = 5*np.exp(1j*np.deg2rad(0))
R = 100
Z_R = R
L = 0.5
Z_L = 1j*w_a*L
# serie impedans
Z = Z_R + Z_L

fig, axs = plt.subplots(figsize=(5,3))
axs.plot(f, abs(Z),'b')
axs.set_xlabel('frekvens [Hz]')
axs.set_ylabel('Impedans |Z| [ohm]')
```
