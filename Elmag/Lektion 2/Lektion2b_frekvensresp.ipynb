## Räknar/ritar frekvensrespons för LR krets




#### Koden nedan räknar och ritar amplituden av totala impedansen i funktion av frekvens i område: 0 – 1000 Hz

### Din uppgift är att räknar och ritar amplituden av strömmen $I$ genom kretsen i funktion av frekvens i område: 0 – 1000 Hz och tolka resultat. Vad händer med strömmen och impedansen?

---

```python
# variera frekvens
f = np.linspace(1,1000,200)
w_a = 2*np.pi*f
V0 = 5*np.exp(1j*np.deg2rad(0))
R = 100
Z_R = R
L = 0.5
Z_L = 1j*w_a*L
# serie impedans
Z = Z_R + Z_L

I = V0/Z

fig1, axs = plt.subplots(1,2, figsize=(10,5))
axs[0].plot(f, abs(Z),'b', label='Impedans')
axs[0].set_xlabel('frekvens [Hz]')
axs[0].set_ylabel('Impedans [Ohm]')
axs[1].plot(f, abs(I))
axs[1].set_xlabel('frekvens [Hz]')
axs[1].set_ylabel('Ström [A]')
```

Eftersom vi har en spole där Impedansen är proportionel mot frekvensen så går den upp med frekvensen. Strömen är däremot proportionel mot $ \frac{1}{Z}$ så när Z går upp går strömmen ner.

## Uppgift 2.4



#### Rita i funktion av frekvens i område: 2 kHz – 50 kHz :
- magnituden av totala impedansen 
- amplituden av strömmen
- amplituden av spänningen genom resistor $U_R$


#### Tolka resultat. 

---


```python
f = np.linspace(2000, 50000, 500)
w = 2*np.pi*f
L = 1j*10*10**-3*w
C = 1j*(-1/(10*10**-9*w))
R = 100
V0 = cm.rect(10,0)

Z = L+C+R
I = V0/Z
UR = R*I

fig2, axs2 = plt.subplots(1,3, figsize=(15,3))
axs2[0].plot(f, abs(Z))
axs2[1].plot(f, abs(I))
axs2[2].plot(f, abs(UR))
for ax in axs2:
    ax.set_xlabel('Frekvens [Hz]')
axs2[0].set_ylabel('Impedans [Ohm]')
axs2[1].set_ylabel('Ström [A]')
axs2[2].set_ylabel('Spänning [V]')
```

Vi har ett minimi i Impedans vid ca 1500 Hz detta är alltså kretsens resonans frekvens, här påverkar endast R då reaktanserna från L och C tar utvarandra.
