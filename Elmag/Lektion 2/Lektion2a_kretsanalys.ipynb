```python
# lsning med nod-analys
E1 = 20
E2 = 50
R1 = 1
R2 = 2
R3 = 3
R4 = 4
R5 = 5

Y = np.array([[ -1/R1 - 1/R2 - 1/R3, 1/R2],
        [1/R2, -1/R2 - 1/R4 - 1/R5 ]])

I = np.array([-E1/R1, E2/R5])
V = np.linalg.solve(Y, I)
U1 = V[0]
U2 = V[1]
I2 = (U1-U2)/R2
print('Strmmen I2 rknad med nodanalys r ', round(I2, 3))

# lsning med sling-analys
Z =  np.array([[ R1+R3, -R3, 0],
               [-R3, R2+R3+R4, -R4 ],
                [0, -R4, R4+R5]])

V = np.array([ E1, 0, E2])
I = np.linalg.solve(Z, V)
I2 = I[1]
print('Strmmen I2 rknad med slinganalys r ', round(I2, 3))
```
