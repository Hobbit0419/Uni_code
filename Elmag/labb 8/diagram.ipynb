```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
R = 1010
RL = 94
L = 75.5 * 10**-6
C = 24 *10**-9
f = np.array([1000,2000,3000,4000,5000,6000,8000,10000,3500,4500,3760])
w = 2*np.pi*f

U = np.array([0.62, 1.5, 2.85, 3.48, 2.64, 1.96, 1.24, 0.96, 3.48, 3.04, 3.5])
T = np.array([1000,500,330,252,200,168,126,100,284,224,266]) * 10**-6
dT = np.array([200,410,40,10,24,26,100,22,20,20,20]) * 10**-6

I = U/(np.sqrt((R+RL)**2 + (w*L - (1/(w*C)))**2))*1000

phit = np.arctan((w*L-(1/(w*C)))/(R+RL))
phi = 2*np.pi * dT/T
```

```python
fig,ax = plt.subplots()
ax.scatter(f, phit, label='Teoretisk fasförskjutning')
ax.scatter(f, phi, label='Uppmätt fasförskjutning')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(f, I)
```
