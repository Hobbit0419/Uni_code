## **Växelströmeffekt - effekt anpassning** 

### Maximum effektöverföring

Problemet gäller hur man får ut största möjliga effekt ur ett givet generatorsystem, dvs max utvecklad effekt $P_{max}$ i en yttre belastning $Z_L = R_L + jX_L$.

"Generatorn" (t.ex. utgången från en förstärkare) representeras som spänningskälla  $E_G$ i serie med inre impedans $Z_G = R_G + jX_G$



### Den aktiva effekten kan beräknas för kretsen ovan ur:

\begin{align}
P_{ut} = \text{Re} (\bar{V}\cdot\bar{I^*})= \text{Re} \left( |E_G|^2 \frac{R_L + jX_L}{(R_L+R_G)^2+(X_L+X_G)^2} \right)
\end{align}

### Det kan visas att maximum effektöverföring får man när:

\begin{align}
\bar{Z_L} = \bar{Z_G^*} \qquad  \text{dvs} \quad R_L = R_G \quad \text{och} \quad  X_L = - X_G 
\end{align}

### Tre fall kan förekomma:

1. Du känner till inre impedansen, $Z_G$, och måste välja en belastningsimpedans, $Z_L$, för
maximal effektöverföring så du försöker ställa in den lika med konjugat av $Z_G$.
2. Om du kan bara variera resistiva delen av belastningsimpedans, $R_L$, maximum effektöverföring får du med $R_L = \sqrt{R_G^2 + (X_G+ X_L)^2}$
3. Om du har bara resistiv last, $Z_L = R_L$, maximum effektöverföring får du med $R_L = \sqrt{R_G^2 + X_G^2} = |Z_G|$

---

### Effektfaktorn och Faskompensering:
 
\begin{align}
P &= U \cdot I \cos{(\phi)} \qquad \text{Aktiv effekt}\\
Q &= U \cdot I \sin⁡{(\phi)} \qquad \text{Reaktiv effekt}\\
f_p &= \frac{P}{S} = \cos{\phi} \qquad \text{Effektfaktorn} \quad \text{ (1 för resistiva element)}
\end{align}

#### Med komplexa, $j\omega$-metoden: 

\begin{align}
 \bar{S} = P + jQ = S (\cos{\phi} + j\sin{\phi}) = \bar{U}\cdot\bar{I}^* 
\end{align}

###  Faskompensering

De flesta av de belastningar (motorer, lysrör, etc) som är anslutna till växelströmsnätet har induktiv impedans och kräver således magnetiseringsström eller med andra ord reaktiv effekt.
Man säger att belastningen “förbrukar” reaktiv effekt. Den reaktiva effekten, som ej är någon effekt i egentlig mening, “produceras” i elkraftstationernas generatorer. Den mellan
strömkälla och belastning oscillerande magnetiseringsströmmen (den s.k. reaktiva strömkomponenten) belastar alltså ledningsnätet, varvid onödiga spänningsfall och aktiva effektförluster uppkommer. Det matade ledningsnätet kommer i detta fall att arbeta med $\cos(\phi)<1$.
Genom att till den induktiva belastningen parallellkoppla en kondensator kan man reducera strömmen i nätet ($\cos(\phi)$ ökar. Reaktiv effekt ”pendlar”  utan överföringsförluster och endast den nödvändiga strömmen levereras från generator.
